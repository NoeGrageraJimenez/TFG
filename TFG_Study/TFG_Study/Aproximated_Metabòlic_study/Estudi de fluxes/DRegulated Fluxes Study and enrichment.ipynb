```python
def flux_transformer(v):

    '''
    This function transforms fluxes into log2 scale with three additions:
    - zero values are kept.
    - values between 0 and 2 are adjusted to fit below one.
    - negative fluxes are considered. 
    '''

    def adjusted(v):
        new = (np.log2(v + 0.125) + 3) / 4
        return np.min([new, 1])

    if v == 0:
        vprime = 0
    elif v > 0:
        if v < 2:
            vprime = adjusted(v)
        else:
            vprime = np.log2(v)
    elif v < 0:
        v = np.abs(v)
        if v < 2:
            vprime = -adjusted(v)
        else:
            vprime = -np.log2(v)
    else:
        raise ValueError('found value not considered')

    return vprime
```

```python
t_Cmodel=df_Cmodel.applymap(flux_transformer)
t_Cmodel_T2D= df_Cmodel_T2D.applymap(flux_transformer)
t_Cmodel_T3cD=df_Cmodel_T3cD.applymap(flux_transformer)
t_Cmodel_T3cD
```

```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats import multitest
from statsmodels.stats.multitest import multipletests

t_statistic, p_values = stats.ttest_ind(t_Cmodel_T2D, t_Cmodel, axis = 0, nan_policy = 'omit')
fold_change = np.mean(t_Cmodel_T2D, axis=0) - np.mean(t_Cmodel, axis=0)
fold_change[np.isnan(fold_change)] = 0
p_values[np.isnan(p_values)] = 1.0  
reject,q_values, _, _ = multipletests(
    p_values,
    alpha=0.05,
    method='fdr_bh',  
    is_sorted=False,
    returnsorted=False)

t_results = pd.DataFrame({
    'reaction': t_Cmodel.columns,
    'p_value': p_values,
    'q_value': q_values,
    'Diferencia': fold_change
})

t_results['sign'] = 'NORMAL'
for idx, row in t_results.iterrows():
    if row['q_value'] < 0.05 and row['Diferencia'] > 1:
        t_results.loc[idx, 'sign'] = 'UP'
    elif row['q_value'] < 0.05 and row['Diferencia'] < -1:
        t_results.loc[idx, 'sign'] = 'DOWN'

t_fluxes_sign = t_results[t_results['sign'] != 'NORMAL']
print(t_fluxes_sign)
print(t_fluxes_sign['sign'].to_string(index=False))
```

```python
import seaborn as sns
plt.figure(figsize = (10, 6))
colors = {"NORMAL":"#D2D2D2", "UP":"#FF6462", "DOWN":"#62B8FF"}

sns.scatterplot(x = t_results['Diferencia'], y = -np.log10(t_results['q_value']), hue = t_results['sign'], palette = colors, alpha = 0.7)
plt.axhline(-np.log10(0.05), color = 'k', linestyle = '--', linewidth = 1, label = 'THRESHOLD')
plt.axvline(1, color = 'k', linestyle = '--', linewidth = 1)
plt.axvline(-1, color = 'k', linestyle = '--', linewidth = 1)

plt.xlabel('Diferencia')
plt.ylabel('-log10(q-value)')
plt.title('Reaccions Diferencialment Regulades Cmodel_T2D vs Cmodel')
plt.legend()
plt.grid(True)


up_down_results = t_results[(t_results['q_value'] < 0.05) &
                            ((t_results['sign'] == 'UP') | (t_results['sign'] == 'DOWN'))]
top_10 = up_down_results.nsmallest(10, 'q_value')

# Agregar etiquetas de texto para el nombre de la reacción en los 10 puntos más significativos
for index, row in top_10.iterrows():
    plt.text(row['Diferencia'], -np.log10(row['q_value']), row['reaction'], fontsize=8, ha='center', va='center')

    
plt.show()
```

sf(k, M, n, N, loc=0)
k: El número de éxitos en la muestra.
M: Número total de nombre d'objectes en la poblacion total. (valor númèric de la dimensió)
n: El número de éxitos en la población total, nombre total de objectes Tipus I (random variate .
N: El tamaño de la muestra.
loc: (Opcional) Un parámetro de desplazamiento. Por defecto es 0.

```python
import re
import math
from scipy.stats import hypergeom
#Aplicar Hypergeomètric test per cada un subsystem, la funció de normal només aplica per a un. 
#Per fer-ho  per tots el faré amb loop.
T2D_subsystems = subsystems_df[subsystems_df.index.isin(T2D_results.index)]
k = T2D_subsystems.groupby('Subsystem').size().sort_values(ascending=False)
n = subsystems_df.groupby('Subsystem').size().sort_values(ascending=False)

p_values = {}
enrichment_scores = {}

for subsystem, k_values in k.items():
    # Total number of subsystems
    M = len(subsystems_df)
    
    # Count of occurrences of the subsystem in the entire dataset
    n_value = n.get(subsystem, 0)  # Handle cases where the subsystem is not present in n
    
    # Total number of subsystems in ts_subsystems
    N = len(T2D_subsystems)
    
    # Get the first element of k_value list and convert to integer
    k_value=int(k_values)
    
   
    
    # Enrichment score: observed count / expected count
    enrichment_score = (k_value /N)*100
    
    enrichment_scores[subsystem] = enrichment_score
    # Calculate the p-value using the hypergeometric distribution
    p_value = hypergeom.sf(k_value -1 , M, n_value, N)
    
    # Store the p-value in a dictionary
    p_values[subsystem] = p_value

# Apply FDR correction
q_values = multipletests(list(p_values.values()), method='fdr_bh')[1]

log_q_values = [-math.log10(q) for q in q_values]
# Print or use the p-values as needed


df = pd.DataFrame({
    'Subsystems': list(p_values.keys()),
    'Counts':  [k[subsystem] for subsystem in p_values.keys()],
    'Enrichment_Score': list(enrichment_scores.values()),
    'p_values': list(p_values.values()),
    'q_values': list(q_values)
})
filtered_df = df[df['q_values'] < 0.05]
filtered_df =filtered_df.sort_values(by='Enrichment_Score', ascending=False)



def extract_intermediate(subsystem):
    match = re.search(r'\[<Group (.*?) at 0x', subsystem)
    if match:
        return match.group(1)
    return subsystem

# Aplicar la función a la columna 'Subsystems'
filtered_df['Subsystems'] = filtered_df['Subsystems'].apply(extract_intermediate)
filtered_df
```
