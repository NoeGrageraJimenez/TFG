```python
from cobra.io import load_model
from scipy.io import loadmat
from cobra.sampling import sample
from cobra.io import load_matlab_model
```

```python
Cmodel = load_matlab_model("C:/Users/NOE/Desktop/TFG-Example/Islet data/Genes downregulated/Modified Models/Cmodeln.mat")
Cmodel_T2D = load_matlab_model("C:/Users/NOE/Desktop/TFG-Example/Islet data/Genes downregulated/Modified Models/Cmodel_T2D.mat")
Cmodel_T3cD = load_matlab_model("C:/Users/NOE/Desktop/TFG-Example/Islet data/Genes downregulated/Modified Models/Cmodel_T3cD.mat")
```

```python
co = Cmodel.problem.Constraint(Cmodel.reactions.biomass_maintenance.flux_expression, lb=500)
Cmodel.add_cons_vars([co])
co_T2D = Cmodel_T2D.problem.Constraint(Cmodel_T2D.reactions.biomass_maintenance.flux_expression, lb=500)
Cmodel_T2D.add_cons_vars([co_T2D])
co_T3cD = Cmodel_T3cD.problem.Constraint(Cmodel_T3cD.reactions.biomass_maintenance.flux_expression, lb=500)
Cmodel_T3cD.add_cons_vars([co_T3cD])
```
