# Creaci√≥ DataFrame:

## 1. DataFrame IGT:

```python

import os
import pandas as pd

# Ruta de la carpeta que contiene los archivos .txt
carpetaIGT = "C:/Users/NOE/Desktop/TFG-Example/Islet data/Mostres/Archivos con symbol/IGT"

# Lista para almacenar los DataFrames de los archivos .txt
dataframes = []

# Recorremos todos los archivos en la carpeta
for archivo in os.listdir(carpetaIGT):
    if archivo.endswith(".txt"):
        # Construimos la ruta completa del archivo
        ruta_completa = os.path.join(carpetaIGT, archivo)
        # Cargamos el archivo .txt en un DataFrame
        df = pd.read_csv(ruta_completa, sep='\t', index_col=0)  # Ajusta el separador si es necesario
        # Almacenamos el DataFrame en la lista
        dataframes.append(df)
        print(f"Longitud de {archivo}: {len(df)}")

# Combinamos todos los DataFrames en uno solo
df_completo = pd.concat(dataframes, axis=1)

ruta_guardado = "C:/Users/NOE/Desktop/TFG-Example/Islet data/DataFrames/IGT_DataFrame.csv"
df_completo.to_csv(ruta_guardado)

# Visualizamos el DataFrame completo
df_completo
```

## DataFrame Conjunt:

```python
import os
import pandas as pd

# Definimos las rutas de las carpetas que contienen los archivos .txt
directorios = {
    "IGT": "C:/Users/NOE/Desktop/TFG-Example/Islet data/Mostres/Archivos con symbol/IGT",
    "ND": "C:/Users/NOE/Desktop/TFG-Example/Islet data/Mostres/Archivos con symbol/ND",
    "T2D": "C:/Users/NOE/Desktop/TFG-Example/Islet data/Mostres/Archivos con symbol/T2D",
    "T3cD": "C:/Users/NOE/Desktop/TFG-Example/Islet data/Mostres/Archivos con symbol/T3cD"
}

# Funci√≥n para cargar los DataFrames de los archivos .txt
def cargar_dataframes(directorios):
    dataframes = {}
    for key, carpeta in directorios.items():
        dataframes[key] = []
        # Recorremos todos los archivos en la carpeta
        for archivo in os.listdir(carpeta):
            if archivo.endswith(".txt"):
                # Construimos la ruta completa del archivo
                ruta_completa = os.path.join(carpeta, archivo)
                # Cargamos el archivo .txt en un DataFrame
                df = pd.read_csv(ruta_completa, sep='\t', index_col=0)  # Ajusta el separador si es necesario
                # Almacenamos el DataFrame en la lista
                dataframes[key].append(df)
                print(f"Longitud de {archivo}: {len(df)}")
    return dataframes

# Cargamos todos los DataFrames
dataframes = cargar_dataframes(directorios)

# Concatenamos todos los DataFrames y a√±adimos el nombre correspondiente a cada tipo
df_completo = pd.concat([pd.concat([df.set_axis([f"{key}_{col}" for col in df.columns], axis=1) for df in dfs], axis=1) for key, dfs in dataframes.items()], axis=1)

# Guardamos el DataFrame en un archivo CSV
ruta_guardado = "C:/Users/NOE/Desktop/TFG-Example/Islet data/DataFrames/AllConditions_DataFrame.csv"
df_completo.to_csv(ruta_guardado)

num_columnas_IGT = sum(col.startswith('IGT') for col in df_completo.columns)
num_columnas_ND = sum(col.startswith('ND') for col in df_completo.columns)
num_columnas_T2D = sum(col.startswith('T2D') for col in df_completo.columns)
num_columnas_T3cD = sum(col.startswith('T3cD') for col in df_completo.columns)

print(f"N√∫mero de columnas que comienzan con IGT: {num_columnas_IGT}")
print(f"N√∫mero de columnas que comienzan con ND: {num_columnas_ND}")
print(f"N√∫mero de columnas que comienzan con T2D: {num_columnas_T2D}")
print(f"N√∫mero de columnas que comienzan con T3cD: {num_columnas_T3cD}")

# Visualizamos el DataFrame completo
df_completo
```
